```python
# int()- It converts a value into integer
a=3.7
print(int(a))
# float()-It converts a value into decimal number
b=5
print(float(b))
# bool()- It converts a value into true or false
print(bool(0))
print(bool(5))
#complex-It converts a value into complex form having real and imaginary part
print(complex(2,3))
#str()-changes a value into text format
c=10
print(str(c))
#list()-changes a sequence into list form using square brackets
d=(1,2,3)
print(list(d))
#tuple()-changes a sequence into tuple form using round brackets
e=[6,7,8]
print(tuple(e))
#set()-changes a sequence into set and removes repeated values uses curly brackets
f=[3,4,4,5]
print(set(f))
#dict()-changes data into key and value pair form
g={"name":"pihu","contact":2345,"age":20,'address':'Noida'}
print(g)
#bin()-changes integer into binary representation
print(bin(5))
#oct()-changes integer into octal representation 
print(oct(8))
#hex()-changes integer into hexadecimal representation
print(hex(10))
```
