# Conditional Statements

#### 1. if | indentation

```python
Num=23
```

```python
if Num>0:
    print('Positive')
# 1 when condition in front of if is true then next line of code will execute
# : means indentation means next will start after 4 spaces[indent
# above both line od code will consider as single block
```

#### 2. elif

- elif will execute once the condition of if is false

```python
if Num>0:
    print("Positive")           #jb tk if wala false nhi hoga tb tk elif nhi work krega
elif Num>2:
    print("Positive and Greater than 2")
```

```python
if Num>0:
    print("Positive")
if Num>2:
    print("Positive and Greater than 2")
```

```python
First_number=float(input("Please enter yout first number for calculation"))
operator= input("please enter a operator from given list[+,-,/,*]")
Second_number=float(input("Please enter your second number for calculation"))
if operator == "+":
    print(First_number+Second_number)
elif operator == "-":
    print(First_number-Second_number)
elif operator == "/":
    print(First_number/Second_number)
elif operator == "*":
    print(First_number*Second_number)
```

```python
First_number=float(input("Please enter yout first number for calculation"))
operator= input("please enter a operator from given list[+,-,/,*]")
Second_number=float(input("Please enter your second number for calculation"))
if operator == "+":
    print(First_number+Second_number)
elif operator == "-":
    print(First_number-Second_number)
elif operator == "/":
    try:
      print(First_number/Second_number)
    except:
      print("Number cannot be divided by zero")
elif operator == "*":
    print(First_number*Second_number)
```

#### 3. else

```python
First_number=float(input("Please enter yout first number for calculation"))
operator= input("please enter a operator from given list[+,-,/,*]")
Second_number=float(input("Please enter your second number for calculation"))
if operator == "+":
    print(First_number+Second_number)
elif operator == "-":
    print(First_number-Second_number)
elif operator == "/":
    try:
      print(First_number/Second_number)
    except:
      print("Number cannot be divided by zero")
elif operator == "*":
    print(First_number*Second_number)
else:
    print("This is not a defined operator please enter + |-| / or* only")
#else will execute only when all conditions of both if and elif are false
```
